##### 1.Write python function for creating a random oscialltion with amplitude in range (1,5) and frequency in range (0.25,1)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
sns.set()
```

```python
fs=100
ts=np.arange(0,60,1/fs)
# initialize figure
plt.figure(figsize=[24,12])
#iterate over each amplitude and frequency and create an oscillation plot
for i in range(1,5):
    amp=100*1/float(i)
    freq=0.25*i
    
    ys=amp*np.sin(2*np.pi*freq*ts)
    
    lb="amp:"+str(amp)[0:4]+ ",freq:"+str(freq)[0:4]
    plt.plot(ts,ys, label=lb)

plt.xlabel('Time in Seconds')
plt.ylabel('Displacement')
plt.legend()
plt.show()
```

##### 2. Create 20 different oscillations by implementing python function defined above and implement superposition principle to obtain single resultant oscillation.

```python
fs=100
ts=np.arange(0,60,1/fs)
# initialize figure
plt.figure(figsize=[22,6])
#iterate over each amplitude and frequency and create an oscillation plot

amps=[1*1/float(i)for i in range(1,21)]
freqs=[0.25*i for i in range(1,21)]
    
ys=sum([amp*np.sin(2*np.pi*freq*ts) for amp,freq in zip(amps,freqs)])
    

plt.plot(ts,ys, label='Superposed Oscillation')


plt.xlabel('Time in Seconds', fontsize=22)
plt.ylabel('Displacement', fontsize=22)
plt.legend()
plt.show()
```
